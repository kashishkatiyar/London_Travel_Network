```python
df = df.dropna(how='all')
```

```python
df['Period beginning'] = pd.to_datetime(
    df['Period beginning'],
    format='%d-%b-%y'
)

df['Period ending'] = pd.to_datetime(
    df['Period ending'],
    format='%d-%b-%y'
)
```

```python
df[['Period beginning', 'Period ending']].dtypes
```

```python
df = df.rename(columns={
    'Period and Financial year': 'period_financial_year',
    'Reporting Period': 'reporting_period',
    'Days in period': 'days_in_period',
    'Period beginning': 'period_beginning',
    'Period ending': 'period_ending',
    'Bus journeys (m)': 'bus_journeys',
    'Underground journeys (m)': 'underground_journeys',
    'DLR Journeys (m)': 'dlr_journeys',
    'Tram Journeys (m)': 'tram_journeys',
    'Overground Journeys (m)': 'overground_journeys',
    'Emirates Airline Journeys (m)': 'emirates_airline_journeys',
    'TfL Rail Journeys (m)': 'tfl_rail_journeys'
})
```
